## Estimando os parâmetros da regressão linear simples

Qual será a relação entre a avaliação $y$ e o número de calorias $x$ de um cereal? Para responder esta pergunta, considere uma regressão linear simples
$$
y = \beta_0 + \beta_1 x.
$$
Para encontrar os coeficientes $\beta_0$ e $\beta_1$ utilizando o método dos mínimos quadrados, basta resolver o sistema
$$
\begin{bmatrix}
n & \sum_i x^{(i)} \\
\sum_i x^{(i)} & \sum_i (x^{(i)})^2
\end{bmatrix}
\begin{bmatrix}
\beta_0 \\ \beta_1
\end{bmatrix}
=
\begin{bmatrix}
\sum_i y^{(i)} \\ \sum_i x^{(i)} y^{(i)}
\end{bmatrix}
$$

Portanto, para encontrar $\beta_0$ e $\beta_1$, você precisa
1. Calcular a matriz
$$
A = \begin{bmatrix}
n & \sum_i x^{(i)} \\
\sum_i x^{(i)} & \sum_i (x^{(i)})^2
\end{bmatrix}
$$
e o vetor
$$
c = \begin{bmatrix}
\sum_i y^{(i)} \\ \sum_i x^{(i)} y^{(i)}
\end{bmatrix}
$$
2. Resolver $A \beta = c$, onde $\beta$ é o vetor de coeficientes.

**Exercício 1 - Regressão simples:** Encontre os coeficientes $\beta_0$ e $\beta_1$ quando a variável independente é ```calories```. Dica: A variavel X abaixo já armazena os valores deste atributo.

````python
X = new_df['calories'].values

# insira seu código aqui

A = np.array([[X.shape[0], X.sum()],[X.sum(), X@X.T]])

c = np.array([[y.sum()],[y@X]])

beta = np.linalg.inv(A) @ c
print(beta)
````

**Exercício 2 - Regressão múltipla:** Considerando a nova tabela de dados X abaixo com os atributos 'calories', 'protein', 'fat', 'sugars' e 'vitamins' selecionados, estime os parâmetros da regressão múltipla para obter a variavel resposta ```rating``` 

````python
X = new_df.loc[:,['calories', 'protein', 'fat', 'sugars', 'vitamins']].values

# insira seu código aqui

ones = np.ones(X.shape[0]).reshape(X.shape[0],1)

X = np.concatenate((ones,X), axis=1)

rating = (np.linalg.inv(X.T @ X) @ X.T @ y).reshape(6,1)
print(rating)
````

**Exercício 3:** Nossos modelos de regressão linear são bons preditores da nota de avaliação do cereal? Qual o melhor modelo? Calcule os coeficientes de determinação e faça uma análise dos valores obtidos para responder a estas perguntas.

````python
# insira seu código aqui

delta_y_simples = (y - np.array(new_df['calories'].values*beta[1] + beta[0]))**2

coef_reg_simples = 1 - np.sum(delta_y_simples)/np.sum((y - np.mean(y))**2)

delta_y_multipla = (y.reshape(74,1) - X@rating)**2

coef_reg_multipla = 1 - np.sum(delta_y_multipla)/np.sum((y - np.mean(y))**2)

print("Coeficiente de determinacao da regressao linear simples: " , coef_reg_simples, "\n")
print("Coeficiente de determinacao da regressao linear multipla: " , coef_reg_multipla)
````

Comparando-se os coeficientes de determinação, é possível perceber que a regressão linear simples apresentou um coeficiente bastante distante de 1 (0.481), o que indica que o modelo não é muito bom preditor da nota de avaliação do cereal. Já a regressão linear múltipla apresentou um coeficiente bem mais perto de 1 (0.834). Assim, pode-se concluir que o o modelo obtido pela regressão linear múltipla foi melhor pois ele se ajustou bem aos dados coletados, o que implica em uma predição mais significativa da avalição de um novo cereal de acordo com as suas características nutricionais.
